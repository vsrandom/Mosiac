```python
import cv2
import numpy as np
```

```python
rimg = cv2.imread("./images/2.jpeg")
rimg_gray = cv2.cvtColor(rimg,cv2.COLOR_BGR2GRAY)
# rims=cv2.resize(img1,(960,540))
limg=cv2.imread("./images/1.jpeg")
limg_gray = cv2.cvtColor(limg,cv2.COLOR_BGR2GRAY)
# lims=cv2.resize(img2,(960,540))
sift = cv2.xfeatures2d.SIFT_create()

kp1, des1 = sift.detectAndCompute(rimg_gray,None)
kp2, des2 = sift.detectAndCompute(limg_gray,None)

for d in des1:
    print(d)
```

```python
#Brute force matching
bf=cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(des1,des2)
matches = sorted(matches,key=lambda x:x.distance)
#minimum distance implies good matching !
for m in matches:
    print(m.distance)
```

```python
matching_result = cv2.drawMatches(rimg_gray,kp1,limg_gray,kp2,matches[:20],None,flags=2)
#ims1=cv2.drawKeypoints(ims1,keypoints,None)
cv2.imshow("Matching Result",matching_result)
cv2.imshow("right image",rimg_gray)
cv2.imshow("left image",limg_gray)
k=cv2.waitKey(0)
if k==27:
    cv2.destroyAllWindows()
```

```python
good = []
for m in matches[:20]:
    good.append(m)
#print(good,3*'\n',kp1,4*'\n')
# for m in good :
#     src_points.append(np.float32(kp1[m.queryIdx].pt))
# print(src_points)
# x=src_points.reshape(-1,1,2)
# print(2*'\n',x)
src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)
M,mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0)
print(M)
h,w = rimg_gray.shape
print(h,w)
```

```python
pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
dst=cv2.perspectiveTransform(pts,M)
print(pts,2*'\n',dst)
# lims=cv2.polylines(lims,[np.int32(dst)])
```

```python
limg_gray=cv2.polylines(limg_gray,[np.int32(dst)],True,255,5,cv2.LINE_AA)
cv2.imshow("original image overlapping",limg_gray)
k=cv2.waitKey(0)
if k==27:
    cv2.destroyAllWindows()
```

```python
final_image=cv2.warpPerspective(rimg,M,(limg.shape[1]+rimg.shape[1],limg.shape[0]))
final_image[0:limg.shape[0],0:limg.shape[1]]=limg
print(final_image[0].shape,10*'\n',final_image)
cv2.imshow("final stichhed image",final_image)
k=cv2.waitKey(0)
if k==27:
    cv2.destroyAllWindows()
```
